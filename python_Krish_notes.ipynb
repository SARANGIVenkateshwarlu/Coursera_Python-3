```python
num=24
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
```

```python
## Call this function
even_or_odd(99)
```

### Lambda Function

```python
#Syntax
lambda arguments: expression
```

```python
def addition(a,b):
    return a+b
```

```python
print(addition(4,8))
```

```python
addition = (lambda a,b:a+b)
print(addition(4,8))
```

```python
def even(num):
    if num%2==0:
        return True
    
even(24)
```

```python
even1=(lambda num:num%2==0)
even1(12)
```

```python
def addition(x,y,z):
    return x+y+z

addition(12,13,14)
```

```python
## map()- applies a function to all items in a list
numbers=[1,2,3,4,5,6]
def square(number):
    return number**2

square(6)
```

```python
list(map(lambda x:x**2,numbers))
```
